```python
import pandas as pd
```

```python
TFL = pd.read_csv('data/tfl_journeys_cleaned.csv')
```

```python
%%sql
SELECT 
    YEAR,
    SUM(JOURNEYS_MILLIONS) AS Total_Journeys_Millions
FROM 
    TFL
GROUP BY 
    YEAR
ORDER BY 
    YEAR;
```

```python
%%sql
SELECT 
    YEAR,
    JOURNEY_TYPE,
    SUM(JOURNEYS_MILLIONS) AS Total_Journeys_Millions
FROM 
    TFL
GROUP BY 
    YEAR, JOURNEY_TYPE
ORDER BY 
    YEAR, Total_Journeys_Millions DESC
LIMIT 12;
```

```python
%%sql
SELECT 
    MONTH,
    AVG(JOURNEYS_MILLIONS) AS Average_Journeys_Millions
FROM 
    TFL
GROUP BY 
    MONTH
ORDER BY 
    MONTH;
```
