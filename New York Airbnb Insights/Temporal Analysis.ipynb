```python
import pandas as pd
```

```python
airbnb = pd.read_csv('airbnb_data.csv')
```

```python
%%sql
SELECT 
    DATE_TRUNC('month', CAST(listing_added AS DATE)) AS month_year,
    COUNT(*) AS num_listings
FROM 
    airbnb
GROUP BY 
    month_year
ORDER BY 
    num_listings DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    DATE_TRUNC('month', CAST(listing_added AS DATE)) AS month_year,
    AVG(price) AS avg_price,
    AVG(availability_365) AS avg_availability
FROM 
    airbnb
GROUP BY 
    month_year
ORDER BY 
    month_year DESC, avg_price ASC
LIMIT 10;
```

```python
%%sql
SELECT 
    DATE_TRUNC('month', CAST(listing_added AS DATE)) AS month_year,
    AVG(rating) AS avg_rating
FROM 
    airbnb
WHERE 
    rating IS NOT NULL
GROUP BY 
    month_year
ORDER BY 
    month_year DESC
LIMIT 10;
```

1. **Number of Listings Over Time:**
   - Analyzed the trend of listing additions over different months.
   - Calculated the number of listings added in each month.
   - Identified the months with the highest number of listings added.

2. **Seasonal Trends in Pricing and Availability:**
   - Investigated seasonal patterns in pricing and availability of Airbnb listings.
   - Calculated the average price and availability for each month.
   - Identified months with significant changes in pricing or availability, potentially indicating peak seasons or trends in the market.

3. **Trend of Ratings Over Time:**
   - Explored the trend of average ratings of Airbnb listings over different months.
   - Calculated the average rating for each month.
   - Identified any significant changes or fluctuations in ratings over time.

The results of these analyses provide insights into the temporal dynamics of Airbnb listings, including trends in listing additions, pricing, availability, and ratings over different months. These insights can be valuable for hosts, travelers, and stakeholders in the hospitality industry to understand the seasonal patterns and make informed decisions.
