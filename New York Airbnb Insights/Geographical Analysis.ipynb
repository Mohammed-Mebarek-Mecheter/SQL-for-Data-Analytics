```python
import pandas as pd
```

```python
airbnb = pd.read_csv('airbnb_data.csv')
```

```python
%%sql
SELECT neighbourhood_full, COUNT(*) AS listing_count
FROM airbnb
GROUP BY neighbourhood_full
ORDER BY listing_count DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    neighbourhood_full,
    AVG(price) AS avg_price,
    AVG(rating) AS avg_rating
FROM 
    airbnb
GROUP BY 
    neighbourhood_full
ORDER BY 
    avg_price DESC, avg_rating DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    neighbourhood_full,
    AVG(availability_365) AS avg_availability,
    AVG(number_of_reviews) AS avg_reviews
FROM 
    airbnb
GROUP BY 
    neighbourhood_full
ORDER BY 
    avg_availability DESC, avg_reviews DESC
LIMIT 10;
```

1. **Top Neighborhoods by Number of Listings:**
   - Identified the neighborhoods with the highest number of Airbnb listings.
   - Revealed insights into the distribution of listings across different boroughs and neighborhoods in New York.

2. **Top Neighborhoods by Average Price and Rating:**
   - Explored the average price and rating of listings in each neighborhood.
   - Highlighted neighborhoods with higher average prices and ratings, potentially indicating desirable areas for travelers willing to pay more for better-rated accommodations.

3. **Top Neighborhoods by Availability and Number of Reviews:**
   - Investigated neighborhoods with high availability and number of reviews.
   - Identified neighborhoods where accommodations are consistently available and attract a high number of reviews, suggesting popular destinations or well-received listings.

These analyses provide insights into the geographical distribution, pricing, ratings, availability, and popularity of Airbnb listings across different neighborhoods in New York. They aid in understanding the dynamics of the rental market and help travelers make informed decisions based on their preferences and priorities.
