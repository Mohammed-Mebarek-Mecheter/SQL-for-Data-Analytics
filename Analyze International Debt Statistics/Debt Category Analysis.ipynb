```python
import pandas as pd
```

```python
international_debt = pd.read_csv("international_debt.csv")
```

```python
%%sql
SELECT country_name, indicator_name, SUM(debt) AS total_debt
FROM international_debt
GROUP BY country_name, indicator_name
ORDER BY country_name, total_debt DESC
LIMIT 5;
```

```python
%%sql
SELECT indicator_name, AVG(debt) AS average_debt
FROM international_debt
GROUP BY indicator_name
ORDER BY average_debt DESC;
```
