```python
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv")
```

```python
%%sql
SELECT
    product_line,
    warehouse,
    client_type,
    SUM(quantity) AS total_quantity
FROM
    sales
GROUP BY
    product_line, warehouse, client_type
ORDER BY total_quantity DESC
LIMIT 10;
```

```python
%%sql
SELECT
    CORR(unit_price, quantity) AS correlation_coefficient
FROM
    sales;
```

```python
%%sql
SELECT
    unit_price,
    SUM(quantity) AS total_quantity,
    SUM(total) AS total_revenue
FROM
    sales
GROUP BY
    unit_price
ORDER BY total_revenue DESC
LIMIT 10;
```
