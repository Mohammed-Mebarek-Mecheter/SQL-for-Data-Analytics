```python
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv")
```

```python
%%sql
SELECT
    strftime('%Y-%m', CAST(date AS DATE)) AS month,
    SUM(total) AS total_sales
FROM
    sales
WHERE
    CAST(date AS DATE) BETWEEN '2021-06-01' AND '2021-08-31'
GROUP BY
    month
ORDER BY
    month;
```

```python
%%sql
SELECT
    strftime('%Y-%m', CAST(date AS DATE)) AS month,
    strftime('%W', CAST(date AS DATE)) AS week,
    SUM(total) AS total_sales
FROM
    sales
WHERE
    CAST(date AS DATE) BETWEEN '2021-06-01' AND '2021-08-31'
GROUP BY
    month, week
ORDER BY
    month, week;
```

# Sales Trends Analysis

This report provides an analysis of sales trends based on the provided dataset for the period from June to August 2021.

## Overview

The sales trends analysis aims to understand how sales performed over time during the specified period. By examining the total sales for each month, we can identify patterns, fluctuations, and potential factors influencing sales.

## Findings

### Overall Sales Trends

- Total sales in June 2021 amounted to $95,320.03.
- Total sales in July 2021 amounted to $93,547.91.
- Total sales in August 2021 amounted to $100,245.06.

### Observations

1. **June to July Comparison:**
   - Total sales slightly decreased from June to July, indicating a minor downturn in sales during this period.

2. **July to August Comparison:**
   - Total sales increased from July to August, showing a notable uptick in sales during this period.

3. **Overall Trend:**
   - Despite the decrease in July, the overall trend from June to August is upward, with an increase in total sales over the three-month period.

4. **Possible Factors:**
   - Seasonal influences, marketing campaigns, economic conditions, or changes in consumer behavior could have contributed to the observed fluctuations in sales.
