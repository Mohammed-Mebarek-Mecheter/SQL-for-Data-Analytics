```python
import pandas as pd
```

```python
prouduct_emissions = pd.read_csv('product_emissions_clean.csv')
```

```python
%%sql
SELECT 
    industry_group,
    year,
    SUM(carbon_footprint_pcf) AS total_carbon_footprint
FROM 
    prouduct_emissions
GROUP BY 
    industry_group, year
ORDER BY 
    total_carbon_footprint DESC
LIMIT 10;
```

This query provides valuable insights into the industries with the highest carbon emissions for each year. For example:

In 2015, the "Electrical Equipment and Machinery" industry group had the highest total carbon footprint.
The "Automobiles & Components" industry group appears multiple times in different years, indicating significant carbon emissions across various years.
Other industries such as "Materials," "Food, Beverage & Tobacco," and "Technology Hardware & Equipment" also feature in the top 10 list.

```python
%%sql
SELECT
    industry_group,
    SUM(carbon_footprint_pcf) AS total_emissions
FROM
    prouduct_emissions
GROUP BY
    industry_group
ORDER BY
    total_emissions DESC;
```

Based on the provided data, here's a summary of the comparison:

1. **Electrical Equipment and Machinery**: This industry group has the highest total emissions, indicating a significant carbon footprint associated with the production of electrical equipment and machinery.

2. **Automobiles & Components**: The automotive industry also shows substantial emissions, although slightly lower than electrical equipment and machinery.

3. **Materials, Technology Hardware & Equipment, Capital Goods**: These industries also contribute to emissions, but their total emissions are comparatively lower than the top two industries.

4. **Food, Beverage & Tobacco, Pharmaceuticals, Biotechnology & Life Sciences, Software & Services, Chemicals, Media**: These industries show varying levels of emissions, with food, beverage, and tobacco having relatively higher emissions compared to others.

5. **Energy, Forest and Paper Products, Mining, Consumer Durables & Apparel, Commercial & Professional Services, Containers & Packaging, Tires, Food & Staples Retailing, Consumer Durables, Household and Personal Products, Telecommunication Services, Trading Companies & Distributors, Textiles, Apparel, Footwear and Luxury Goods, Food & Beverage Processing, Utilities, Gas Utilities, Semiconductors & Semiconductor Equipment, Retailing, Tobacco, Household & Personal Products**: These industries exhibit relatively lower total emissions compared to the top industries.

Based on this comparison, industries such as Electrical Equipment and Machinery and Automobiles & Components may have the greatest potential for emission reduction initiatives due to their significant carbon emissions. However, emissions reduction efforts should be considered across all industries to mitigate overall environmental impact.

```python
%%sql
SELECT 
    SUM(carbon_footprint_pcf) AS total_emissions,
    SUM(carbon_footprint_pcf * upstream_percent_total_pcf / 100) AS production_emissions,
    SUM(carbon_footprint_pcf * operations_percent_total_pcf / 100) AS transportation_emissions,
    SUM(carbon_footprint_pcf * downstream_percent_total_pcf / 100) AS disposal_emissions
FROM 
    prouduct_emissions;
```

The query results indicate the total emissions as well as the contributions of each stage (production, transportation, disposal) to the overall emissions:

- **Total Emissions**: The total emissions from all products in the dataset amount to approximately 13.7 million carbon dioxide equivalent units.
- **Production Emissions**: Production activities contribute around 5.77 million carbon dioxide equivalent units to the total emissions.
- **Transportation Emissions**: Transportation contributes about 3.06 million carbon dioxide equivalent units to the total emissions.
- **Disposal Emissions**: Disposal activities contribute roughly 4.86 million carbon dioxide equivalent units to the total emissions.

These results provide insights into the distribution of emissions across different stages of the product lifecycle, highlighting the significant contributions of production and disposal activities.
